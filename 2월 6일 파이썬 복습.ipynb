```python
# 소득 누적
income = 0

# 자판기 메뉴별 재고 현황
drink = {"coffee" : 10, "cola" : 5, "cider" : 3}

# 자판기 메뉴별 가격
price = {"coffee" : 500, "cola" : 400, "cider" : 300}

def sale(menu, income):
    income = stock_menu_check(menu, "coffee", 1, "커피", income)
        
    income = stock_menu_check(menu, "cola", 2, "콜라", income)
        
    income = stock_menu_check(menu, "cider", 3, "사이다", income)
            
    drink_stock_show()
    return income

# 재고, 음료 확인 후 음료 판매
def stock_menu_check(menu, menu_name, menu_num, menu_name_korean, income):
    
    if stock_check(drink, menu_name) == False:
        print(f'{menu_name_korean}가 매진 되었습니다. 돈을 반환합니다.')
    
    if stock_check(drink, menu_name) and menu == menu_num:
        if money >= price[menu_name]:
            income += price[menu_name]
            drink[menu_name] -= 1
            print(menu_name_korean + "가 나왔습니다. 맛있게 드세요~")
            print(f'잔돈 반환 : {money - price[menu_name]}')
        else:
            print("돈이 부족합니다.")
        
    return income

# 재고 확인
def stock_check(drink, drink_name):
    if drink[drink_name] >= 1:
        return True
    else:
        return False
    
# 재고 현황 보기
def drink_stock_show():
    coffee = drink["coffee"]
    cola = drink["cola"]
    cider = drink["cider"]
    
    drink_name = ["커피", "콜라", "사이다"]
    drink_out = ["매진", "매진", "매진"]
    
    print()
    print("재고 현황")
    if coffee >= 1:
        drink_out[0] = coffee
    print(drink_name[0], drink_out[0])
        
    if cola >= 1:
        drink_out[1] = cola
    print(drink_name[1], drink_out[1])
        
    if cider >= 1:
        drink_out[2] = cider
    print(drink_name[2], drink_out[2])

# 프로그램 시작
print("자판기 프로그램 시작!!")
print()

while True:
    # 화면에 자판기 메뉴 보여주기
    print("메뉴 : Coffee(500원)[1], Cola(400원)[2], Cider(300원)[3]")
    print()

    # 돈과 음료를 입력받기
    money = int(input("돈을 넣으세요. : "))
    menu = int(input("음료를 선택하세요. : 1(Coffee), 2(Cola), 3(Cider), 4(종료) : "))
    print()
    # 음료 확인하고 입력받은 돈이 음료의 가격과 같거나 많다면 음료 판매
    income = sale(menu, income)
    print()
    print("소득 현황 :", income)
    print()
    
    if menu == 4:
        break

print("자판기 프로그램 종료!!")
```
