```python
# 음료 자판기 프로그램
# 메뉴 : 커피(10), 콜라(5), 사이다(3)
# 가격 : 커피(500원), 콜라(300원), 사이다(200원)
# 판매 : 메뉴에 해당하는 돈을 받고
#        상품가격과 돈이 맞는지 비교
#         상품에 맞는 돈이면 상품 판매한다.(상품의 현재 수량을 -1)
# in_money : 판매금액

# 판매금액
in_money = 0

drink = {"커피" : 10, "콜라" : 5, "사이다" : 3}

# 커피, 콜라, 사이다의 가격
price = [500, 300, 200]

def sale(menu, money, in_money):
    if count_check(drink, "커피") and menu == "1":
        if money >= price[0]:
            in_money += price[0]
            drink["커피"] -= 1
            print("커피 나왔어요 맛있게 드세요.")
            print("잔돈 반환 :", money-price[0])
        else:
            print("돈이 모자랍니다.")
        
    if count_check(drink, "콜라") and menu == "2":
        if money >= price[1]:
            in_money += price[1]
            drink["콜라"] -= 1
            print("콜라 나왔어요 맛있게 드세요.")
            print("잔돈 반환 :", money-price[1])
        else:
            print("돈이 모자랍니다.")

    if count_check(drink, "사이다") and menu == "3":
        if money >= price[2]:
            in_money += price[2]
            drink["사이다"] -= 1
            print("사이다 나왔어요 맛있게 드세요.")
            print("잔돈 반환 :", money-price[2])
        else:
            print("돈이 모자랍니다.")
    return in_money

# 재고수량 체크
def count_check(drink, drink_name):
    if drink[drink_name] >= 1:
        return True
    else:
        return False
            
# 프로그램 시작

while True:
    # 화면에 자판기를 보여준다
    print("메뉴 : 커피(500원)[1], 콜라(300원)[2], 사이다(200원)[3]")

    # 사용자로부터 돈을 입력 받는다
    money = int(input("돈을 입력 : "))
    menu = input("음료를 선택 1(커피), 2(콜라), 3(사이다), 4(종료) : ")

    # 입력한 돈을 확인하여 상품을 판매할 수 있는지 결정
    in_money = sale(menu, money, in_money)
    print("상품재고 :", drink)
    print("판매금액 :", in_money)
    
    if menu == "4":
        break
        
print("음료 자판기 프로그램 종료되었습니다.")
```
